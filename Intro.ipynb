```python
print("Hello, Kevin!") # This line prints the text "Hello, Python!" to the console.
```

```python
# Assign a string (text) to a variable named 'name'
name = "Kevin"

# Assign an integer (whole number) to a variable named 'age'
age = 14

# Print the values stored in the variables
print("My name is", name)
print("I am", age, "years old")
```

Let's perform some basic arithmetic operations:

```python
# Addition
sum_result = 10 + 5
print(f"10 + 5 = {sum_result}")

# Subtraction
diff_result = 20 - 7
print(f"20 - 7 = {diff_result}")

# Multiplication
prod_result = 4 * 6
print(f"4 * 6 = {prod_result}")

# Division
div_result = 25 / 5
print(f"25 / 5 = {div_result}")

# Exponentiation (power)
pow_result = 2 ** 3 # 2 to the power of 3
print(f"2 ** 3 = {pow_result}")

# Modulo (remainder of division)
mod_result = 10 % 3
print(f"10 % 3 = {mod_result}")
```

Let's explore some common Python data types:

```python
# Integer (whole number)
my_integer = 10
print(f"my_integer: {my_integer}, Type: {type(my_integer)}")

# Float (number with decimal)
my_float = 3.14
print(f"my_float: {my_float}, Type: {type(my_float)}")

# String (text)
my_string = "Hello, Python Learner!"
print(f"my_string: {my_string}, Type: {type(my_string)}")

# Boolean (True or False)
my_boolean = True
print(f"my_boolean: {my_boolean}, Type: {type(my_boolean)}")
```
